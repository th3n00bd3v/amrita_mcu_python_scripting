```python
# make a plot of histograms
# which ones got created? why?

import pandas as pd

# Load data

cols = ['age',
'workclass',
'fnlwgt',
'education',
'education-num',
'marital-status',
'occupation',
'relationship',
'race',
'sex',
'capital-gain',
'capital-loss',
'hours-per-week',
'native-country',
'income']
df = pd.read_csv('adult.csv', names = cols)


histogram = df.hist()
```

```python
# adjust the bins

bins_adjusted = df.hist(bins=5)
```

```python
# zoom on in the outliers of 'capital gain'

capital_gain = df.hist(column="capital-gain", bins=5, range=(0, 5000))
```

```python
# make a correlation plot (look up how we did it in the demo notebook)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

numeric_data = df.select_dtypes(include=['number'])
corr_matrix = numeric_data.corr()
print(corr_matrix)

pd.plotting.scatter_matrix(numeric_data, figsize=(12, 12), diagonal='kde')
plt.suptitle("Scatter Matrix of Numeric Features")
plt.show()
```
