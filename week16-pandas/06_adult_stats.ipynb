# load the adult dataset up again

```python
# What is the average value for fnlwgt?
import pandas as pd
import numpy as np

cols = ['age',
'workclass',
'fnlwgt',
'education',
'education-num',
'marital-status',
'occupation',
'relationship',
'race',
'sex',
'capital-gain',
'capital-loss',
'hours-per-week',
'native-country',
'income']
df = pd.read_csv('adult.csv', names = cols)

average_fnlwgt = df['fnlwgt'].mean()
print(average_fnlwgt)
```

```python
# if you sum the values of each row, which row has the largest values?

numeric_cols = df.select_dtypes(include='number')
sum_rows = numeric_cols.sum(axis=1)
largest_sum_index = sum_rows.idxmax()
print("Row with largest sum is at index:", largest_sum_index)

# view the full row
print(df.loc[largest_sum_index])
```

```python
#which row has the smallest values? (if summed)

numeric_cols = df.select_dtypes(include='number')
sum_rows = numeric_cols.sum(axis=1)
smallest_sum_index = sum_rows.idxmin()
print("Row with smallest sum is at index:", smallest_sum_index)

# view the full row
print(df.loc[smallest_sum_index])
```

```python
# What is the standard deviation of Capital gain?

capital_gain_std = df['capital-gain'].std()
print(capital_gain_std)
```

```python
# what is the average of the difference between capital-gain and capital-loss?
# (find the difference of the two columns and then take the average)

avg_diff_capital_gainloss = (df['capital-gain'] - df['capital-loss']).mean()
print(avg_diff_capital_gainloss)
```

```python
# insert some missing values into your dataframe 
# select 10 rows and overwrite them with np.nans

missing_indices = np.random.choice(df.index, size=10, replace=False)
df.loc[missing_indices] = np.nan

print(df.loc[missing_indices])
```

```python
# count the number of na's in your dataset

nan_counts = df.isna().sum()
print(nan_counts)
```
