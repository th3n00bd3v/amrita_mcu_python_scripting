# load the adult dataset

```python
# use the code .value_counts() on a column to see all
# unique values for that column
# for example
# df['workclass'].value_counts() should show you all the 
# possible unique values in workclass

import pandas as pd
import numpy as np

cols = ['age',
'workclass',
'fnlwgt',
'education',
'education-num',
'marital-status',
'occupation',
'relationship',
'race',
'sex',
'capital-gain',
'capital-loss',
'hours-per-week',
'native-country',
'income']
df = pd.read_csv('adult.csv', names = cols)

df['education'].value_counts()
```

```python
# create a boolean series for the column sex
# then use it to select all rows where the person is a women

df['sex'] = df['sex'].str.strip()


is_woman = df['sex'] == 'Female'
women = df[is_woman]
print(women.head())
```

```python
# select all rows where the eductaion is bachelors or HS-Grad
    
df['education'] = df['education'].str.strip()

education_bachelors = (df['education'] == 'Bachelors') | (df['education'] == 'HS-grad')
education_bachelors_or_hsgrad = df[education_bachelors]

print(education_bachelors_or_hsgrad.head())
```

```python
#select all rows where education-num is greater than 12

education_num = df['education-num'] > 12
education_num_greater_than_12 = df[education_num]
print(education_num_greater_than_12.head())
```

```python
# select the two columns capital-gain and capital-loss
# where the work class is either Private or State-gov

df['workclass'] = df['workclass'].str.strip()

capitalgainorloss = (df['workclass'] == 'Private') | (df['workclass'] == 'State-gov')
capitalgainloss = df[capitalgainorloss][['capital-gain', 'capital-loss']]
print(capitalgainloss.head())
```

```python
# select all rows of the dominant relationship type 
# (whichever one has the majority number of rows)
# or if they are divorced

# so basically
# figure out what the most common relationship entry is, call it X
# then select all rows where X is true OR they are divorced (from marital status)

X = df['relationship'].value_counts().idxmax()
print(X)

condition = (df['relationship'] == X) | (df['marital-status'] == 'Divorced')
result = df[condition]
print(result.head())
```
